## Sentence and Word tokenization

```python
import spacy
```

```python
# Using english package
# We create an object for that
nlp = spacy.load("en_core_web_sm")
```

```python
#There is 2 sentences here
doc = nlp("We love God and we try to do our best to be a better person.God bless you")
```

```python
# Let's split the sentences
# It's easier to write ode this way(ORP)
for sentence in doc.sents:
    print(sentence)
# You can see that the sentences are split now
```

```python
# Separating words

for sentence in doc.sents:
    for word in sentence :
        print(word)
```

```python
# So many options here
from nltk.tokenize import sent_tokenize
```

```python
sent_tokenize("We love God and we try to do our best to be a better person.God bless you")
```

```python
from nltk.tokenize import word_tokenize
```

```python
word_tokenize("We love God and we try to do our best to be a better person.God bless you")
```
