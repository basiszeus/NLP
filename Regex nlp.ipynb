```python
chat1 = "you ask a lot of questions, there is my phone 1234567892 and my email : abc@xyz.com"
chat2 = "here it is : 1234567892,abc@xyz.com"
chat3 = "Tired of you  asking thing !!! there is my phone (123)-456-7894, and my email :abc@xyz.com"
```

```python
pattern = '\d{10}'
matche_1 = re.findall(pattern, chat1)
matche_1
```

```python
match_2 = re.findall(pattern, chat2)
match_2
```

## How about the 3rd chat ?

- We need to find digits inside bracket, for that we can use regex101 to find a way. We can write 

Then we will add the rest

# How do we match both pattern at the same time ?

We need to add 'Or' expression, we will use '|'

```python
pattern = '\d{9}|\(\d{3}\)-\d{3}-\d{4}'
match_3 = re.findall(pattern, chat3)
match_3
```

# Now Emails 

```python
#you can write [a-z0-9]*@ if you have digits also
#you can write [a-z0-9A-Z]*@ if you have caps too
#you can write [a-z0-9A-Z_]*@ if you have underscore also
```

```python
pattern = '[a-z]*@[a-z]*.[a-z]*'
match = re.findall(pattern, chat3)
match
```

# Order number

```python
message4 = 'there is my order # 123456789'

pattern = '# \d{9}'

match = re.findall(pattern,message4)
match
```

Another way

```python
# We have order + # + sequence of digits
# To find what is after order ; "like order number" we must match everything except a digits, to do that we can use [^\d]* 
# Finally we write \d* to find the sequence of character for the code number
```

```python
#If you want to identify only a portion , you need to use brackets
#Now you have your match and the order number in a group
```

```python
message4 = 'there is my order number # 123456789'

pattern = 'order[^\d]*(\d*)'

match = re.findall(pattern,message4)
match
```
