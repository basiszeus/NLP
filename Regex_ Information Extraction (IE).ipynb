We will use Regex to find key information on specific person, we will use a primitive approche , there is a modern one like Google do, but it's just for the sack of understanding what we can do with Regex

```python
import re
```

We will take info from this page : https://en.wikipedia.org/wiki/Muhammad

```python
Text = '''
Born 570 CE (53 BH)[1]
Mecca, Hejaz, Arabia
Died	June 8, 632 (11 AH) (aged 61–62) Medina, Hejaz, Arabia
Resting place	Green Dome at al-Masjid an-Nabawi, Medina, Arabia 24°28′03″N 39°36′41″E
Spouse	See Muhammad's wives
Children	See Muhammad's children
Parents	Abdullah ibn Abd al-Muttalib (father)Amina bint Wahb (mother)
Known for	Founding Islam
Other names	
Rasūl Allāh (Messenger of God)(see Names and titles of Muhammad)
Relatives	Family tree of Muhammad, Ahl al-Bayt  ("Family of the House")
Arabic name
Personal (Ism)	Muḥammad
Patronymic (Nasab)	Muḥammad ibn ʿAbd Allāh ibn ʿAbd al-Muṭṭālib ibn Hāshim ibn ʿAbd Manāf ibn Quṣayy ibn Kilāb
Teknonymic (Kunya)	ʾAbu al-Qāsim
Epithet (Laqab)	Ḵhātam an-Nabiyyīn (Seal of the Prophets)'''
```

# Age

For Age we can do : aged.(\d*.\d*) 

```python
pattern = 'aged.(\d*.\d*)'
match = re.findall(pattern, Text)
match
```

# Date of death

For Date of death we can do : Died.([A-Za-z]*.\d,.\d*)

```python
pattern = 'Died.([A-Za-z]*.\d,.\d*)'
match = re.findall(pattern, Text)
match
```

# Known for what ?

For Kown for what you can write : Known for.([A-Za-z]*.[A-Za-z]*)


```python
pattern = 'Known for.([A-Za-z]*.[A-Za-z]*)'
match = re.findall(pattern, Text)
match[0]
#When you do [0] you remove the "[]" in the output
```

# Birth Place

```python
# \n mean the line bellow
# So we will take everything in the line bellow Born
```

For "Birth Place" you can write : Born.*\n(.*)


```python
pattern = 'Born.*\n(.*)'
match = re.findall(pattern, Text)
match[0]
```

# Date of Birth

For Kown for what you can write : Born(.*)\CE

```python
# "\" means when you should stop
# You can also write it like this
pattern = 'Born.(\d*.[A-Z]*)'
match = re.findall(pattern, Text)
match[0]
```
